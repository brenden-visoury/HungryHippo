```python
import pandas as pd #data
import requests as rq
import json
```

```python
df = pd.read_csv('../HungryHippo/apikeys.csv')
zwsid =  df.loc[df['API'] == 'zillow-api']['KEY'].iloc[0]
rapid_id = df.loc[df['API'] == 'rapid-api']['KEY'].iloc[0]
```

```python
import requests

url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"

querystring = {"location":"28217","status_type":"RecentlySold","home_type":"Houses","sort":"Price_High_Low","bedsMin":"1"}

headers = {
    'x-rapidapi-key': rapid_id,
    'x-rapidapi-host': "zillow-com1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

# print(response.text)

results = response.text
```

```python
results_json = json.loads(results)
```
